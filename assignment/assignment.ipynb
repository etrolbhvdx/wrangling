**Q1.** Open the "tidy_data.pdf" document in the repo, which is a paper called Tidy Data by Hadley Wickham.

  1. Read the abstract. What is this paper about?

  This paper specifically focuses on the "data tidying" aspect of data cleaning. The paper discusses the structure that tidy data should follow and explores how following tidy data principles can make exploring data easier.
  2. Read the introduction. What is the "tidy data standard" intended to accomplish?

  The tidy data standard is meant to give an easy framework that data scientists can follow to prepare their datasets. Having a standard provides direction and makes the process of data cleaning less painful.
  3. Read the intro to section 2. What does this sentence mean: "Like families, tidy datasets are all alike but every messy dataset is messy in its own way." What does this sentence mean: "For a given dataset, itâ€™s usually easy to figure out what are observations and what are variables, but it is surprisingly difficult to precisely define variables and observations in general."

  The first sentence means that since tidy datasets follow a certain framework, they all contain the same underlying structure and are approachable and easy to work with. Messy datasets, on the other hand, could have any number of problems that make them difficult to work with, since they were not created with the idea of clean data in mind. The second sentence refers to the idea that when working with a specific dataset, it can be clear to us what each part of the data represents.
  4. Read Section 2.2. How does Wickham define values, variables, and observations?

  Wickham defines values as specific numbers, strings, etc. belonging to a variable and an observation (like an individual cell). A variable contains all values that measure the same attribute (like a column). An observation is a specific measurement of a value (a row).
  5. How is "Tidy Data" defined in section 2.3?

  In section 2.3, tidy data is listed as having three conditions: each variable forms a column, each observation forms a row, and each type of observational unit forms a table.
  6. Read the intro to Section 3 and Section 3.1. What are the 5 most common problems with messy datasets? Why are the data in Table 4 messy? What is "melting" a dataset?
  
  The five most common problems with messy datasets are the following: column headers are values instead of variable names, multiple variables are stored in one column, variables are stored in both rows and columns, multiple types of observational units are stored in the same table, and a single observational unit is stored in multiple tables. The data in table 4 is messy because it does not follow the conditions for tidy data: both the rows and column headers contain values (like agnostic or '<'10k) instead of variables (like religion) and observations (specific measurements of these variables). Melting refers to the process of turning columns into rows, specifically columns that do not represent variables, and making the table tidy.
  7. Why, specifically, is table 11 messy but table 12 tidy and "molten"?

  Table 11 has specific days as columns, when these are individual values. Table 12 fixes this by having values of days in a date column. In addition, tmax and tmin measure different things (min and max temperature) and should be considered variables, with the values being temperature measurements. Table 12 also fixes this issue.
  8. Read Section 6. What is the "chicken-and-egg" problem with focusing on tidy data? What does Wickham hope happens in the future with further work on the subject of data wrangling?

  Section 6 lists the "chicken-and-egg" problem as the following: if tidy data is only as useful as the tools that work with it,then tidy tools will be inextricably linked to tidy data. Wickham is suggesting that sometimes, we get too stuck in the weeds of fixing a specific tool or dataset that we are unable to improve data cleanliness practices as a whole. Wickham hopes that we can build on this framework of tidy data and continue to make more broadly applicable frameworks that streamline the process of data cleaning.

```python
#2 Categorical
sharks=pd.read_csv('./data/sharks.csv')
sharks['Type'].value_counts()
```

```python

sharks['Type']=sharks['Type'].replace(['Questionable','Unconfirmed','Unverified','Under investigation'],'Invalid')
sharks['Type']=sharks['Type'].replace(['Boat','Boatomg','Watercraft','Boating'],'Boating')
sharks['Type'].value_counts()
```

```python
#3: Dummy Variable
#I downloaded this data online, but the command below could load it directly from the website
#url = 'http://www.vcsc.virginia.gov/pretrialdataproject/October%202017%20Cohort_Virginia%20Pretrial%20Data%20Project_Deidentified%20FINAL%20Update_10272021.csv'
#pretrial = pd.read_csv(url,low_memory=False)
pretrial=pd.read_csv('./data/2017PretrialData.csv')
pretrial['WhetherDefendantWasReleasedPretrial'].value_counts()
```

```python
pretrial['WhetherDefendantWasReleasedPretrial']=pretrial['WhetherDefendantWasReleasedPretrial'].replace(9,np.nan)
pretrial['WhetherDefendantWasReleasedPretrial'].value_counts()
pretrial['WhetherDefendantWasReleasedPretrial'].isna().sum()
```

```python
#4: Missing values, not at random
pretrial['ImposedSentenceAllChargeInContactEvent']=pd.to_numeric(pretrial['ImposedSentenceAllChargeInContactEvent'],errors='coerce')
pretrial['ImposedSentenceAllChargeInContactEvent'].isna().sum()
```

```python
pretrial['SentenceTypeAllChargesAtConvictionInContactEvent'].value_counts()
```

```python
pd.crosstab(pretrial['ImposedSentenceAllChargeInContactEvent'].isna(),pretrial['SentenceTypeAllChargesAtConvictionInContactEvent'])
```

```python
pretrial['ImposedSentenceAllChargeInContactEvent']=np.where(pretrial['SentenceTypeAllChargesAtConvictionInContactEvent']==4,0,pretrial['ImposedSentenceAllChargeInContactEvent'])
pretrial['ImposedSentenceAllChargeInContactEvent']=np.where(pretrial['SentenceTypeAllChargesAtConvictionInContactEvent']==9,np.nan,pretrial['ImposedSentenceAllChargeInContactEvent'])
pd.crosstab(pretrial['ImposedSentenceAllChargeInContactEvent'].isna(),pretrial['SentenceTypeAllChargesAtConvictionInContactEvent'])
```
